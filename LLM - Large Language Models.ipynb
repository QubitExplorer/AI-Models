# # Stock Data Analysis and LLM Summary

```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from transformers import pipeline

# Function to calculate RSI
def calculate_rsi(data, window=14):
    delta = data['Close'].diff(1)
    gain = delta.where(delta > 0, 0)
    loss = -delta.where(delta < 0, 0)

    avg_gain = gain.rolling(window=window, min_periods=1).mean()
    avg_loss = loss.rolling(window=window, min_periods=1).mean()

    rs = avg_gain / avg_loss
    rsi = 100 - (100 / (1 + rs))
    return rsi

# Function to calculate VWAP
def calculate_vwap(data):
    vwap = (data['Close'] * data['Volume']).cumsum() / data['Volume'].cumsum()
    return vwap

# Function to perform live stock analysis
def live_stock_analysis(stock_symbol, model_name='distilgpt2'):
    # Fetch stock data from Yahoo Finance
    stock_data = yf.download(stock_symbol, period='5d', interval='1h')

    # Ensure there is enough data to proceed
    if stock_data.empty:
        print(f"No data found for {stock_symbol}. Please check the stock symbol or try again later.")
        return

    # Drop rows with NaN values to avoid indexing issues
    stock_data.dropna(inplace=True)
    if stock_data.empty:
        print(f"Not enough data after cleaning for {stock_symbol}. Cannot perform analysis.")
        return

    # Calculate RSI and VWAP
    stock_data['RSI'] = calculate_rsi(stock_data)
    stock_data['VWAP'] = calculate_vwap(stock_data)

    # Plotting the stock closing price, RSI, and VWAP
    fig, ax = plt.subplots(3, 1, figsize=(5, 5))

    ax[0].plot(stock_data.index, stock_data['Close'], label='Close Price', color='blue')
    ax[0].set_title(f'{stock_symbol} Close Price')
    ax[0].set_ylabel('Price')
    ax[0].legend()

    ax[1].plot(stock_data.index, stock_data['RSI'], label='RSI', color='red')
    ax[1].set_title('Relative Strength Index (RSI)')
    ax[1].set_ylabel('RSI')
    ax[1].axhline(70, color='gray', linestyle='--')
    ax[1].axhline(30, color='gray', linestyle='--')
    ax[1].legend()

    ax[2].plot(stock_data.index, stock_data['VWAP'], label='VWAP', color='green')
    ax[2].set_title('Volume Weighted Average Price (VWAP)')
    ax[2].set_ylabel('VWAP')
    ax[2].legend()

    plt.tight_layout()
    plt.show()

    # Prepare data for LLM
    if len(stock_data) > 0:
        analysis_text = f"The latest closing price of {stock_symbol} is {float(stock_data['Close'].iloc[-1]):.2f}. " \
                        f"The RSI is {float(stock_data['RSI'].iloc[-1]):.2f}, suggesting {'overbought' if stock_data['RSI'].iloc[-1] > 70 else 'oversold' if stock_data['RSI'].iloc[-1] < 30 else 'neutral'} conditions. " \
                        f"The VWAP is {float(stock_data['VWAP'].iloc[-1]):.2f}."
    else:
        print(f"Not enough data to generate analysis for {stock_symbol}")
        return

    # Load the LLM model
    llm_pipeline = pipeline('text-generation', model=model_name, trust_remote_code=True)

    # Generate analysis summary using LLM
    generated_analysis = llm_pipeline(analysis_text, max_length=100, truncation=True)[0]['generated_text']

    print("\nGenerated Stock Analysis Summary:\n")
    print(generated_analysis)

# Example usage
if __name__ == "__main__":
    live_stock_analysis('AAPL', model_name='distilgpt2')
```

# Weather Data Analysis and LLM Summary

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from transformers import pipeline
import random

# Function to generate sample weather data
def generate_weather_data(days=7, hours_per_day=24):
    date_range = pd.date_range(end=pd.Timestamp.now(), periods=days * hours_per_day, freq='H')
    data = {
        'Temperature': [random.uniform(15, 35) for _ in range(len(date_range))],  # Temperature in Celsius
        'Humidity': [random.uniform(30, 90) for _ in range(len(date_range))],      # Humidity in percentage
        'WindSpeed': [random.uniform(0, 20) for _ in range(len(date_range))]       # Wind speed in km/h
    }
    weather_data = pd.DataFrame(data, index=date_range)
    return weather_data

# Function to perform weather analysis
def weather_analysis(model_name='distilgpt2'):
    # Generate sample weather data
    weather_data = generate_weather_data()

    # Calculate key metrics: average temperature, humidity, wind speed
    avg_temperature = weather_data['Temperature'].mean()
    avg_humidity = weather_data['Humidity'].mean()
    avg_wind_speed = weather_data['WindSpeed'].mean()

    # Plotting the weather data
    fig, ax = plt.subplots(3, 1, figsize=(8, 6))

    ax[0].plot(weather_data.index, weather_data['Temperature'], label='Temperature', color='blue')
    ax[0].set_title('Temperature Over Time')
    ax[0].set_ylabel('Temperature (C)')
    ax[0].legend()

    ax[1].plot(weather_data.index, weather_data['Humidity'], label='Humidity', color='red')
    ax[1].set_title('Humidity Over Time')
    ax[1].set_ylabel('Humidity (%)')
    ax[1].legend()

    ax[2].plot(weather_data.index, weather_data['WindSpeed'], label='Wind Speed', color='green')
    ax[2].set_title('Wind Speed Over Time')
    ax[2].set_ylabel('Wind Speed (km/h)')
    ax[2].legend()

    plt.tight_layout()
    plt.show()

    # Prepare data for LLM
    analysis_text = f"The average temperature for the past week is {avg_temperature:.2f} degrees Celsius. " \
                    f"The average humidity is {avg_humidity:.2f}%, indicating {'high' if avg_humidity > 70 else 'moderate' if avg_humidity > 40 else 'low'} humidity levels. " \
                    f"The average wind speed is {avg_wind_speed:.2f} km/h."

    # Load the LLM model
    llm_pipeline = pipeline('text-generation', model=model_name, trust_remote_code=True)

    # Generate analysis summary using LLM
    generated_analysis = llm_pipeline(analysis_text, max_length=100, truncation=True)[0]['generated_text']

    print("\nGenerated Weather Analysis Summary:\n")
    print(generated_analysis)

# Example usage
if __name__ == "__main__":
    weather_analysis(model_name='distilgpt2')
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from transformers import pipeline
import kaggle
import os
import zipfile

# Download Kaggle dataset
kaggle.api.authenticate()

# Set up Kaggle dataset information
kaggle_dataset = 'new-york-city/nyc-taxi-trip-data'
dataset_zip_path = 'nyc_taxi_trip_data.zip'
dataset_folder = 'nyc_taxi_trip_data'

# Download the dataset
if not os.path.exists(dataset_zip_path):
    kaggle.api.dataset_download_files(kaggle_dataset, path='.', unzip=False)

# Unzip dataset if not already unzipped
if not os.path.exists(dataset_folder):
    with zipfile.ZipFile(dataset_zip_path, 'r') as zip_ref:
        zip_ref.extractall(dataset_folder)

# Load the retail sales data
retail_data = pd.read_csv(f'{dataset_folder}/yellow_tripdata_2019-01.csv')  # Modify the path if necessary

# Function to perform NYC taxi trip analysis
def nyc_taxi_trip_analysis(model_name='distilgpt2'):
    # Calculate key metrics: average sales amount, number of customers, number of transactions
    avg_trip_distance = retail_data['trip_distance'].mean()
    avg_fare_amount = retail_data['fare_amount'].mean()
    avg_trip_duration = (pd.to_datetime(retail_data['tpep_dropoff_datetime']) - pd.to_datetime(retail_data['tpep_pickup_datetime'])).mean().total_seconds() / 60

    # Plotting the retail data
    fig, ax = plt.subplots(3, 1, figsize=(10, 10))

    ax[0].hist(retail_data['trip_distance'], bins=50, label='Trip Distance', color='blue', alpha=0.7)
    ax[0].set_title('Trip Distance Distribution')
    ax[0].set_ylabel('Frequency')
    ax[0].legend()

    ax[1].hist(retail_data['fare_amount'], bins=50, label='Fare Amount', color='red', alpha=0.7)
    ax[1].set_title('Fare Amount Distribution')
    ax[1].set_ylabel('Frequency')
    ax[1].legend()

    ax[2].hist((pd.to_datetime(retail_data['tpep_dropoff_datetime']) - pd.to_datetime(retail_data['tpep_pickup_datetime'])).dt.total_seconds() / 60, bins=50, label='Trip Duration (minutes)', color='green', alpha=0.7)
    ax[2].set_title('Trip Duration Distribution')
    ax[2].set_ylabel('Frequency')
    ax[2].legend()

    plt.tight_layout()
    plt.show()

    # Prepare data for LLM
    analysis_text = f"The average trip distance is {avg_trip_distance:.2f} miles. " " " \
                    f"The average fare amount is ${avg_fare_amount:.2f}. " " " \
                    f"The average trip duration is {avg_trip_duration:.2f} minutes."""

    # Load the LLM model
    llm_pipeline = pipeline('text-generation', model=model_name, trust_remote_code=True)

    # Generate analysis summary using LLM
    generated_analysis = llm_pipeline(analysis_text, max_length=100, truncation=True)[0]['generated_text']

    print("\nGenerated Retail Sales Analysis Summary:\n")
    print(generated_analysis)

# Example usage
if __name__ == "__main__":
    nyc_taxi_trip_analysis(model_name='distilgpt2')
```
