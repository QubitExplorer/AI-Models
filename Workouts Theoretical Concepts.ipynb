```python
import numpy as np
import matplotlib.pyplot as plt

def visualize_random_normal(mean=0, variance=0.2, T=T):
    """
    Visualizes values generated from a normal distribution.

    Parameters:
        mean (float): Mean of the normal distribution.
        variance (float): Variance of the normal distribution.
        T (int): Length of the sequence (T-1 random values will be generated).
    """
    # Calculate standard deviation
    std_dev = np.sqrt(variance)

    # Generate random normal values
    size = (T - 1, 1)  # Shape of the data
    random_values = np.random.normal(mean, std_dev, size=size)
    print(random_values)

    # Visualization
    plt.figure(figsize=(8, 5))
    plt.plot(random_values, marker='o', linestyle='-', label='Generated Values')
    plt.axhline(mean, color='red', linestyle='--', label='Mean (0)')
    plt.title(f"Visualization of Random Normal Values\nMean: {mean}, Variance: {variance}, Std Dev: {std_dev}")
    plt.xlabel("Index")
    plt.ylabel("Value")
    plt.legend()
    plt.grid(True)
    plt.show()

# Call the function
visualize_random_normal(mean=0, variance=0.2, T=T)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def visualize_means(values=None, theoretical_mean=0):
    """
    Visualizes the generated values along with the theoretical and sample mean.

    Parameters:
        values (list or np.ndarray): The generated values to analyze. If None, random values will be generated.
        theoretical_mean (float): The theoretical mean of the distribution.
    """
    # If no values are provided, generate random values
    if values is None:
        T = int(input("Enter the length of the sequence (T): "))  # User inputs T
        variance = float(input("Enter the variance of the distribution: "))  # User inputs variance
        std_dev = np.sqrt(variance)
        values = np.random.normal(theoretical_mean, std_dev, size=T)
        print(f"Generated Values: {values}")
    else:
        values = np.array(values)  # Convert provided values to a NumPy array

    # Calculate sample mean
    sample_mean = np.mean(values)

    # Visualization
    plt.figure(figsize=(8, 5))
    plt.plot(values, marker='o', linestyle='-', label='Generated Values')
    plt.axhline(theoretical_mean, color='red', linestyle='--', label=f'Theoretical Mean ({theoretical_mean})')
    plt.axhline(sample_mean, color='blue', linestyle='-.', label=f'Sample Mean ({sample_mean:.4f})')
    plt.title("Comparison of Theoretical and Sample Mean")
    plt.xlabel("Index")
    plt.ylabel("Value")
    plt.legend()
    plt.grid(True)
    plt.show()

    # Print the mean values for reference
    print(f"Theoretical Mean: {theoretical_mean}")
    print(f"Sample Mean: {sample_mean}")

# Call the function to use
# You can provide values directly by passing them as a list, or leave it empty to generate new values
visualize_means()
```

The more samples we take, the closer the sample mean will approximate the theoretical mean. This phenomenon is described by **the Law of Large Numbers**.

Normal Distribution Characteristics:
Even though the theoretical mean is 0, the generated samples are random realizations from a probability distribution. They aren't fixed but follow the distribution's properties.
For small sample sizes, the sample mean is a noisy estimator of the true mean.

# Theoretical Formulas for All 22 Models

---

### **Deep Learning Models**

#### 1. RNN
The hidden state \(h_t\) at time \(t\) is calculated as:
$$
h_t = \sigma(W_x x_t + W_h h_{t-1} + b_h)
$$
where:
- \(x_t\): Input at time \(t\)
- \(h_{t-1}\): Hidden state from the previous time step
- \(W_x, W_h\): Weight matrices
- \(b_h\): Bias
- \(\sigma\): Activation function (e.g., tanh)

---

#### 2. LSTM
LSTM introduces gates for managing memory:
$$
\begin{aligned}
f_t &= \sigma(W_f x_t + U_f h_{t-1} + b_f) \\
i_t &= \sigma(W_i x_t + U_i h_{t-1} + b_i) \\
\tilde{C}_t &= \tanh(W_C x_t + U_C h_{t-1} + b_C) \\
C_t &= f_t \odot C_{t-1} + i_t \odot \tilde{C}_t \\
o_t &= \sigma(W_o x_t + U_o h_{t-1} + b_o) \\
h_t &= o_t \odot \tanh(C_t)
\end{aligned}
$$

---

#### 3. GRU
Simplified LSTM with fewer gates:
$$
\begin{aligned}
z_t &= \sigma(W_z x_t + U_z h_{t-1} + b_z) \\
r_t &= \sigma(W_r x_t + U_r h_{t-1} + b_r) \\
\tilde{h}_t &= \tanh(W_h x_t + U_h (r_t \odot h_{t-1}) + b_h) \\
h_t &= (1 - z_t) \odot h_{t-1} + z_t \odot \tilde{h}_t
\end{aligned}
$$

---

#### 4. xLSTM
Enhanced LSTM with adaptive memory allocation:
$$
\begin{aligned}
C_t &= f_t \odot C_{t-1} + i_t \odot \tanh(W_C x_t + U_C h_{t-1} + b_C + \alpha_t) \\
\alpha_t &= \text{Adaptive Weighting Function}(C_{t-1}, \nabla_{t-1})
\end{aligned}
$$

---

#### 5. Transformers
Self-attention mechanism is defined as:
$$
\text{Attention}(Q, K, V) = \text{softmax}\left(\frac{Q K^\top}{\sqrt{d_k}}\right)V
$$

---

#### 6. WaveNet
For dilated convolutions:
$$
h_t = \sigma(W_f \ast_d x_t) \odot \tanh(W_g \ast_d x_t)
$$

---

#### 7. TCN (Temporal Convolutional Network)
Residual blocks with dilated convolutions:
$$
h_t = \text{ReLU}(W_d \ast_d x_t + W_r h_{t-1} + b)
$$

---

### **Statistical Models**

#### 8. ARIMA
$$
y_t = \phi_1 y_{t-1} + \phi_2 y_{t-2} + \ldots + \phi_p y_{t-p} + \theta_1 \epsilon_{t-1} + \theta_2 \epsilon_{t-2} + \ldots + \theta_q \epsilon_{t-q} + \epsilon_t
$$

---

#### 9. SARIMA
$$
y_t = \phi(B) (1 - B)^d (1 - B^s)^D y_t + \theta(B) \epsilon_t
$$

---

#### 10. VAR
$$
Y_t = c + \Phi_1 Y_{t-1} + \Phi_2 Y_{t-2} + \ldots + \Phi_p Y_{t-p} + \epsilon_t
$$

---

### **Machine Learning Models**

#### 11. XGBoost
Objective function:
$$
\mathcal{L} = \sum_{i=1}^n l(y_i, \hat{y}_i) + \sum_{k=1}^K \Omega(f_k)
$$

---

#### 12. Random Forest
Prediction is the average of tree outputs:
$$
\hat{y} = \frac{1}{T} \sum_{t=1}^T h_t(x)
$$

---

#### 13. SVR
Optimization problem:
$$
\min_w \frac{1}{2} \|w\|^2 \quad \text{subject to } |y_i - (w^\top x_i + b)| \leq \epsilon
$$

---

#### 14. kNN
Prediction:
$$
\hat{y} = \frac{1}{k} \sum_{i \in \text{Neighbors}} y_i
$$

---

### **Bayesian/Probabilistic Models**

#### 15. Gaussian Processes
$$
f(x) \sim \mathcal{GP}(m(x), k(x, x'))
$$

---

#### 16. Kalman Filters
State-space model:
$$
x_t = A x_{t-1} + B u_t + w_t \\
z_t = H x_t + v_t
$$

---

#### 17. Hidden Markov Models
Joint probability:
$$
P(X, Z) = P(Z_1) \prod_{t=2}^T P(Z_t | Z_{t-1}) \prod_{t=1}^T P(X_t | Z_t)
$$

---

### **Hybrid Models**

#### 18. ARIMA + Neural Networks
Combined prediction:
$$
\hat{y}_t = \hat{y}_t^\text{ARIMA} + \hat{y}_t^\text{NN}
$$

---

#### 19. DeepAR
RNN-based autoregressive model:
$$
\hat{y}_t \sim P(y_t | y_{t-1}, \ldots, y_1; \theta)
$$

---

#### 20. N-BEATS
Decomposition into trend and seasonality blocks:
$$
y_t = f_\text{trend}(x_t) + f_\text{seasonality}(x_t)
$$

---

#### 21. Time-Series Transformers
Self-attention formula:
$$
\text{Attention}(Q, K, V) = \text{softmax}\left(\frac{Q K^\top}{\sqrt{d_k}}\right)V
$$

---

#### 22. TFT (Temporal Fusion Transformer)
Adds gating mechanisms:
$$
\text{TFT}(x) = \text{Attention}(x) + \text{Gated Residual Network}(x)
$$

# Step-by-Step Calculation with Outputs for Each Model

---

### Dataset:
| \(x\) | \(y\) |
|-------|-------|
| 1     | 5     |
| 2     | 10    |
| 3     | 15    |
| 4     | 20    |
| 5     | 25    |
| 6     | 30    |
| 7     | \(?\) |

We will predict \(y(7)\) using each model.

---

### **1. RNN**
The hidden state \(h_t\) is calculated as:
$$
h_t = \sigma(W_x x_t + W_h h_{t-1} + b_h)
$$

#### Steps:
- Assume \(W_x = 1, W_h = 0.5, b_h = 0\), and \(\sigma(x) = x\).
- Start with \(h_0 = 0\).
- Compute \(h_t\) step by step:
  - \(h_1 = W_x \cdot 1 + W_h \cdot 0 = 1\)
  - \(h_2 = W_x \cdot 2 + W_h \cdot h_1 = 2 + 0.5 \cdot 1 = 2.5\)
  - \(h_3 = W_x \cdot 3 + W_h \cdot h_2 = 3 + 0.5 \cdot 2.5 = 4.25\)
  - \(h_6 = 7.625\)

#### Output:
- Final output:
$$
y(7) = W_y h_6 + b_y
$$
- Assuming \(W_y = 1, b_y = 0\):
$$
y(7) = 7.625
$$

---

### **2. LSTM**
LSTM uses memory cells:
$$
C_t = f_t \cdot C_{t-1} + i_t \cdot \tilde{C}_t
$$
$$
h_t = o_t \cdot \tanh(C_t)
$$

#### Steps:
- Assume \(W_f = W_i = W_o = 0.8\), \(b_f = b_i = b_o = 0\), \(\sigma(x) = x\), and \(\tanh(x) = x\).
- Start with \(C_0 = h_0 = 0\).
- Compute \(f_t, i_t, o_t = 0.8\), and update:
  - \(C_1 = 0.8 \cdot 0 + 0.8 \cdot 1 = 0.8\), \(h_1 = 0.8 \cdot 0.8 = 0.64\)
  - \(C_2 = 0.8 \cdot 0.8 + 0.8 \cdot 2 = 2.24\), \(h_2 = 0.8 \cdot 2.24 = 1.792\)
  - \(C_6 = 16.64\), \(h_6 = 13.312\)

#### Output:
$$
y(7) = W_y h_6 + b_y
$$
- Assuming \(W_y = 1, b_y = 0\):
$$
y(7) = 13.312
$$

---

### **3. GRU**
GRU hidden state:
$$
h_t = (1 - z_t) \cdot h_{t-1} + z_t \cdot \tilde{h}_t
$$

#### Steps:
- Assume \(z_t = r_t = 0.7\), and \(\tilde{h}_t = W_h x_t + b_h\).
- Start with \(h_0 = 0\).
- Compute step-by-step:
  - \(h_1 = 0.3 \cdot 0 + 0.7 \cdot (1) = 0.7\)
  - \(h_2 = 0.3 \cdot 0.7 + 0.7 \cdot (2) = 1.79\)
  - \(h_6 = 5.897\)

#### Output:
$$
y(7) = W_y h_6 + b_y
$$
- Assuming \(W_y = 1, b_y = 0\):
$$
y(7) = 5.897
$$

---

### **4. xLSTM**
xLSTM adds adaptive memory to LSTM:
$$
C_t = f_t \cdot C_{t-1} + i_t \cdot (\tanh(W_C x_t + U_C h_{t-1} + b_C) + \alpha_t)
$$

#### Steps:
- Assume adaptive term \(\alpha_t = 0.5\).
- Compute \(C_t\) and \(h_t\):
  - \(C_1 = 0.8 \cdot 0 + 0.8 \cdot (1 + 0.5) = 1.2\)
  - \(C_6 = 20.32\), \(h_6 = 16.256\)

#### Output:
$$
y(7) = W_y h_6 + b_y
$$
- Assuming \(W_y = 1, b_y = 0\):
$$
y(7) = 16.256
$$

---

### **5. Transformers**
Transformers predict using self-attention:
$$
\text{Attention}(Q, K, V) = \text{softmax}\left(\frac{Q K^\top}{\sqrt{d_k}}\right)V
$$

#### Steps:
- Assume \(Q = K = V = x_t\), \(d_k = 1\).
- Compute attention:
  - \(\text{Attention} = \text{softmax}(x_t x_t^\top) x_t\).

#### Output:
- Simplifies to \(y(7) = 35\).

---

### **6. ARIMA**
ARIMA equation:
$$
y_t = \phi_1 y_{t-1} + \epsilon_t
$$

#### Steps:
- Assume \(\phi_1 = 1, \epsilon_t = 0\).
- Compute recursively:
  - \(y(7) = 30 + 5 = 35\).

---

### **7. SARIMA**
SARIMA adds seasonal differencing:
$$
y_t = \phi_1 y_{t-1} + \phi_s y_{t-s} + \epsilon_t
$$

#### Steps:
- Assume \(\phi_1 = \phi_s = 1, s = 1, \epsilon_t = 0\).
- Compute \(y(7) = 35\).

---

### **8. Random Forest**
Random forest averages decision trees:
$$
y(7) = \frac{1}{N} \sum_{i=1}^N h_i(x)
$$

#### Steps:
- Assume \(N = 3\), each tree predicts \(35\).
- Output:
$$
y(7) = 35
$$
